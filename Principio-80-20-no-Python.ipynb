```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
# Atualização do tamanho da fonte do gráfico
plt.rcParams.update({'font.size':14})
```

```python
# Criação do Dataframe
df = pd.DataFrame({'tipo defeito':['nao selagem topo','nao selagem fundo', 'nao selagem lateral', 'impressao borrada', 'caixa amassada', 'outros'], 'freq (%)':[55,38,22,8,7,6]})
df
```

```python
# Coluna com valores acumulados
df['cum (%)'] = df['freq (%)'].cumsum()/df['freq (%)'].sum()*100
df
```

```python
# Gráfico de Pareto com 2 eixos
fig,ax1 = plt.subplots(figsize = (12,6))

ax1.set_title('Pareto')
color1 = "royalblue"
#ax1.set_xlabel('X')
ax1.set_ylabel('Frequência (%)',color = color1)

ax1.bar(df['tipo defeito'], df['freq (%)'],color = color1,edgecolor = 'royalblue',linewidth = 2,\
       hatch = '*')
#ax1.set_ylim([-10,10])
ax1.tick_params(axis = 'y',labelcolor = color1)

color2 = 'black'
ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('%',color = color2)

ax2.plot(df['tipo defeito'], df['cum (%)'],color = color2,marker = 's',markersize = 8, linestyle = '-')

ax2.tick_params(axis = 'y',labelcolor = color2)
ax2.set_ylim([0,120])

for tick in ax1.get_xticklabels():
    tick.set_rotation(45)

plt.savefig('GraficoPareto.png',format='png',dpi = 600, bbox_inches = 'tight')
plt.show()
```
